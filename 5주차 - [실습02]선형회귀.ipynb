
<h1 id="1.-집값-데이터-읽기">1. 집값 데이터 읽기<a class="anchor-link" href="#1.-집값-데이터-읽기">¶</a></h1>


```python
import numpy as np

# (집 크기(평)), (가격(억원))
# 학교근처(홍은동,홍제동,구기동,신영동,평창동,녹번동)

import pandas as pd
data = pd.read_csv('house_price.txt', names=['size','price'])
print(data)

X= data['size'] # 제곱미터(평)
y= data['price'] # 억원
m = len(data)    # 정보 개수(행 개수)
```

```python
# numpy array 형태로 변환, 형태 변환(m) -> (m,1)
X = (np.array(X)).reshape(m,1)
y = (np.array(y)).reshape(m,1)
print(X.shape, y.shape)
```


<h1 id="2.그래프-그리기">2.그래프 그리기<a class="anchor-link" href="#2.그래프-그리기">¶</a></h1>


```python
import matplotlib.pyplot as plt
plt.plot(X,y,'b.')
plt.xlabel("Size of a house in Pyeong") # 집 크기(평)
plt.ylabel("Price in uck-won")          # 매매가(억원)
plt.show()
```


<h1 id="3.-Gradient-descent--&gt;-theta">3. Gradient descent -&gt; theta<a class="anchor-link" href="#3.-Gradient-descent--&gt;-theta">¶</a></h1>


```python
X_b = np.c_[np.ones((m,1)),X] # 모든 샘플에 x0=1을 추가
# c_ : concatenation. 배열을 옆으로 붙이기

learning_rate = 0.0001 # 학습률(learning rate)
n_iter = 200
# 튜닝 해줘야 하는 값들 (learnign rate 와 반복 횟수) 은 하이퍼 파라미터가 된다

theta = np.zeros((2,1))
gradients = np.zeros((2,1))

for i in range(n_iter):
    m = float(m)
    gradients =1/m* X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients
print('theta:')
print(theta)
```


<h1 id="4.-집-크기-최소(10평)~최대(90평)*theta--&gt;-그래프-그리기">4. 집 크기 최소(10평)~최대(90평)*theta -&gt; 그래프 그리기<a class="anchor-link" href="#4.-집-크기-최소(10평)~최대(90평)*theta--&gt;-그래프-그리기">¶</a></h1>


```python
X_new = np.array([[10],[90]]) # 10~90평
X_new_b = np.c_[np.ones((2,1)),X_new] # 모든 샘플에 x0=1을 추가
y_predict = X_new_b.dot(theta)
print("10평과 90평 집의 예측 가격")
print(y_predict)
```

```python
plt.plot(X_new, y_predict,'r-')
plt.plot(X,y,'b.')
plt.xlabel("Size of a house in pyoeng") # 집 크기(10평)
plt.ylabel("Price in uck-won") # 매매가(억원)
plt.show()
```


<h1 id="5.-임의의-집-크기(x_my)에-대한-집값-예측하기">5. 임의의 집 크기(x_my)에 대한 집값 예측하기<a class="anchor-link" href="#5.-임의의-집-크기(x_my)에-대한-집값-예측하기">¶</a></h1>


```python
X_mine = np.array([[25]]) # 우리집이 25평이라면, 얼마일까?
X_mine_b = np.c_[np.ones((1,1)),X_mine] # 모든 샘플에 x0=1을 추가
y_predict = X_mine_b.dot(theta)
print(y_predict, "억원")
```


<h1 id="6.(Optional)'3.-Gradient-descent'대신,-확률적-경사-하강법-함수-호출하여-theta-구하기">6.(Optional)'3. Gradient descent'대신, 확률적 경사 하강법 함수 호출하여 theta 구하기<a class="anchor-link" href="#6.(Optional)'3.-Gradient-descent'대신,-확률적-경사-하강법-함수-호출하여-theta-구하기">¶</a></h1>


```python
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(n_iter=100, penalty=None, eta0=0.0001) # eta0(learning rate)
sgd_reg.fit(X,y.ravel()) #sgd_reg.fit(X, y.ravel()) # 1차원 배열로 만들기 (m,1) -> (m)
```

```python
print("theta:")
print(sgd_reg.intercept_, sgd_reg.coef_)

y_pred = sgd_reg.predict(X)
plt.plot(X,y,'b.')
plt.plot(X,y_pred,'r')
plt.show()
```

```python
#우리집이 25평이라면, 얼마일까?
y_pred = sgd_reg.predict([[25]])
print(y_pred, "억원")
```


<h1 id="7.(Optional)-비용함수-J(Θ)-3차원-그래프로-그리기">7.(Optional) 비용함수 J(Θ) 3차원 그래프로 그리기<a class="anchor-link" href="#7.(Optional)-비용함수-J(Θ)-3차원-그래프로-그리기">¶</a></h1>


```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ComputeJ(t0,t1,X,y):
    m = len(X)
    X_b = np.c_[np.ones((m,1)),X] #모든 샘플에 x0=1을 추가
    theta = np.array([t0, t1])
    
    cost_vector = X_b.dot(theta) -y # (52,1)
    
    cost_vector = cost_vector.reshape(-1)
    # 각 원소 제곱
    cost2 = np.square(cost_vector)
    #원소들 더하기
    cost_sum = np.sum(cost2) / (2*m)
    
    return cost_sum
```

```python
# 1. theta0, theta1, J 값 구하기
d = 100 # # of split

s=1.0
theta0 = np.linspace(-15*s,15*s,d) #-0.1 ~ 0.1
theta1 = np.linspace(-1.0*s,1.0*s,d) #-0.1 ~ 0.2

J = np.ones((d,d)) # (100,100)
for i,t0 in enumerate(theta0):
    for j,t1 in enumerate(theta1):
        cost = ComputeJ(t0,t1,X,y)
        J[i,j] = cost
```

```python
# 2. 그리기
pX, pY = np.meshgrid(theta0,theta1)

fig =plt.figure()
ax = plt.axes(projection ='3d')
ax.contour3D(pX,pY,J,300, cmap='viridis')
ax.set_xlabel('theta0')
ax.set_ylabel('theta1')
ax.set_zlabel('J');
plt.show()
```
