Gaussian kernel 그리기<a class="anchor-link" href="#1.-Gaussian-kernel-그리기">¶</a></h1><h1 id="$-f_1-=-exp(-\frac{||x-l^{(l)}||^2}{2\sigma^2})-$">$ f_1 = exp(-\frac{||x-l^{(l)}||^2}{2\sigma^2}) $<a class="anchor-link" href="#$-f_1-=-exp(-\frac{||x-l^{(l)}||^2}{2\sigma^2})-$">¶</a></h1><h1 id="$-f_1-=-exp(-\frac{||v-1-v_2||^2}{2\sigma^2})-$">$ f_1 = exp(-\frac{||v-1-v_2||^2}{2\sigma^2}) $<a class="anchor-link" href="#$-f_1-=-exp(-\frac{||v-1-v_2||^2}{2\sigma^2})-$">¶</a></h1>


```python

import numpy as np
import matplotlib.pyplot as plt

```

```python

# vector v1과 v2 의 차이에 대한 Gaussian kernal 을 구하자.
# val = GaussianKernal(np.array([0,0]), np.array([3,5]), 1.0)
def GaussianKernal(v1, v2, sigma):
    res= np.exp(-np.power(np.linalg.norm(v1-v2),2)/(2*np.power(sigma,2)))
    
    return res

```

```python

# 힌트
print(np.power([2,3],2))
print(np.sum([2,3]))

val = GaussianKernal(np.array([0,0]), np.array([3,5]), 1.0)
print (val)

```

```python

# 1. x1, x2, Z 값 구하기
d = 100 # # of split
x1 = np.linspace(0,6,d) # 0~6
x2 = np.linspace(0,10,d) # 0~10
l1 = [3,5]

Z = np.zeros((d,d)) # (100,100)
sigma = 1.5 # 1, 0.5(좁게), 1.5(넓게)
for i, t1 in enumerate(x1):
    for j,t2 in enumerate(x2):
        val = GaussianKernal(np.array([t1,t2]), l1, sigma)
        Z[i,j] = val

```

```python

# 2. 그리기

from mpl_toolkits import mplot3d

pX, pY = np.meshgrid(x1, x2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(pX, pY, Z, 100, cmap='jet')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Z')
plt.show()

#ax.view_init(60,35)
fig

```


<h1 id="2.-선형-SVM-분류">2. 선형 SVM 분류<a class="anchor-link" href="#2.-선형-SVM-분류">¶</a></h1><h2 id="1)-데이터-읽기">1) 데이터 읽기<a class="anchor-link" href="#1)-데이터-읽기">¶</a></h2>


```python

import numpy as np

# (x1), (x2), (0 or 1)
# Ng, Machine Learning, Coursera, m1-ex6 중
from scipy.io import loadmat
data = loadmat('data1.mat')
print(data)

X = data['X'] # x1, x2의 값
y = data['y'] # 0 or 1
m = len(y)

```

```python

print(X.shape, y.shape)
```


<h2 id="2)-그래프-그리기">2) 그래프 그리기<a class="anchor-link" href="#2)-그래프-그리기">¶</a></h2>


```python

# y=1, y=0 데이터 인덱스 찾기
pos = []
neg = []

for(i,val) in enumerate(y):
    if val==1:
        pos.append(i)
    else:
        neg.append(i)
        
print(pos)
print(neg)

```

```python

import matplotlib.pyplot as plt
# X[:,1].reshape(-1) : 한 줄로 피기. (m,) -> (m)
plt.plot(X[pos,0].reshape(-1), X[pos,1].reshape(-1), 'b+', label='y=1')
plt.plot(X[neg,0].reshape(-1), X[neg,1].reshape(-1), 'ro', label='y=0')
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='upper right')
plt.title("Data 1")
plt.show()

```


<h2 id="3)-학습">3) 학습<a class="anchor-link" href="#3)-학습">¶</a></h2>


```python

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
# 평균0, 표준편차1 로 만듦 <feature normalization>
from sklearn.svm import LinearSVC
# Linear Support Vector Classification(Linear Support Vector Machine)

svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("linear_svc", LinearSVC(C=1, loss="hinge"))
    # C=1(규제 강화->무뎌지게), C=100(규제 약화->예민하게)
])

svm_clf.fit(X,y)

```

```python

svm_clf.predict([[2.3, 2.7],
                [3.7, 3.9]])

```


<h2 id="4)-Decision-boundary">4) Decision boundary<a class="anchor-link" href="#4)-Decision-boundary">¶</a></h2>


```python

# Plot the decision booundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X[:,0].min()-0.2, X[:,0].max()+0.2
y_min, y_max = X[:,1].min()-0.2, X[:,1].max()+0.2
h = .02 # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max,h))
Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.plot(X[pos,0].reshape(-1), X[pos,1].reshape(-1), 'b+', label='y=1')
plt.plot(X[neg,0].reshape(-1), X[neg,1].reshape(-1), 'ro', label='y=0')
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='upper right')

plt.xlim(xx.min(), xx.max())
plt.xlim(yy.min(), yy.max())
plt.xticks(())
plt.xticks(())

plt.show()

```

<h2 id="1)-데이터-읽기">1) 데이터 읽기<a class="anchor-link" href="#1)-데이터-읽기">¶</a></h2>


```python

import numpy as np

# (x1), (x2), (0 or 1)
# Ng, Machine Learning, Coursera, m1-ex6 중
from scipy.io import loadmat
data = loadmat('data2.mat')
print(data)

X = data['X'] # x1, x2의 값
y = data['y'] # 0 or 1
m = len(y)

```

```python

print(X.shape, y.shape)
```


<h2 id="2)-그래프-그리기">2) 그래프 그리기<a class="anchor-link" href="#2)-그래프-그리기">¶</a></h2>


```python

# y=1, y=0 데이터 인덱스 찾기
pos = []
neg = []

for(i,val) in enumerate(y):
    if val==1:
        pos.append(i)
    else:
        neg.append(i)
        
print(pos)
print(neg)
```

```python

import matplotlib.pyplot as plt
plt.plot(X[pos,0].reshape(-1), X[pos,1].reshape(-1), 'b+', label='y=1')
plt.plot(X[neg,0].reshape(-1), X[neg,1].reshape(-1), 'ro', label='y=0')
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='upper right')
plt.title("Data 2")
plt.show()

```


<h2 id="3)-학습">3) 학습<a class="anchor-link" href="#3)-학습">¶</a></h2>


```python

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
# 평균0, 표준편차1 로 만듦 <feature normalization>
from sklearn.svm import SVC
# Support Vector Classification(Support Vector Machine)

svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("linear_svc", SVC(kernel="rbf", gamma=10, C=1))
    # * kernel: "rbf", "poly", ...
    # * C=1(규제 강화->무뎌지게), C=100(규제 약화->예민하게)
    # * gamma: 1/sigma.
    # 크면 가우시안분포 좁아짐 -> 샘플 영향 범위 작아짐 -> 불규칙&샘플따라 구불구불 -> 예민
    # 작으면 가우시안 분포 넓어짐 -> 샘플 영향 범위 넓어짐 -> smooth -> 무뎌짐
])

svm_clf.fit(X,y)

```


<h2 id="4)-Decision-boundary">4) Decision boundary<a class="anchor-link" href="#4)-Decision-boundary">¶</a></h2>


```python

# Plot the decision booundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X[:,0].min()-0.2, X[:,0].max()+0.2
y_min, y_max = X[:,1].min()-0.2, X[:,1].max()+0.2
h = .02 # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max,h))
Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.plot(X[pos,0].reshape(-1), X[pos,1].reshape(-1), 'b+', label='y=1')
plt.plot(X[neg,0].reshape(-1), X[neg,1].reshape(-1), 'ro', label='y=0')
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='upper right')

plt.xlim(xx.min(), xx.max())
plt.xlim(yy.min(), yy.max())
plt.xticks(())
plt.xticks(())

plt.show()

```
